**Heart Disease Diagnostic Analysis**



```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the heart disease diagnostic database
heart_data = pd.read_csv('/content/Heart Disease data.csv')

# ETL: _Extract, Transform, Load data
# Assuming data is already clean, we proceed to analysis

# Exploratory Data Analysis (EDA)
# Summary statistics
print(heart_data.describe())

# Distribution of heart disease rates
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=heart_data)
plt.title('Distribution of Heart Disease')
plt.xlabel('Target (0: No Heart Disease, 1: Heart Disease)')
plt.ylabel('Count')
plt.show()

# Heart disease by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='target', data=heart_data)
plt.title('Heart Disease by Gender')
plt.xlabel('Gender (0: Female, 1: Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Heart disease by age
plt.figure(figsize=(10, 6))
sns.histplot(x='age', hue='target', data=heart_data, bins=20, kde=True)
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(heart_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Dashboard
# Additional visualizations and analysis can be included in the dashboard

```

```python
# Dashboard

# Distribution of chest pain types
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', hue='target', data=heart_data)
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Distribution of fasting blood sugar levels
plt.figure(figsize=(8, 6))
sns.countplot(x='fbs', hue='target', data=heart_data)
plt.title('Distribution of Fasting Blood Sugar Levels')
plt.xlabel('Fasting Blood Sugar (> 120 mg/dl)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Distribution of resting electrocardiographic results
plt.figure(figsize=(8, 6))
sns.countplot(x='restecg', hue='target', data=heart_data)
plt.title('Distribution of Resting Electrocardiographic Results')
plt.xlabel('Resting Electrocardiographic Results')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Distribution of exercise induced angina
plt.figure(figsize=(8, 6))
sns.countplot(x='exang', hue='target', data=heart_data)
plt.title('Distribution of Exercise Induced Angina')
plt.xlabel('Exercise Induced Angina')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Distribution of number of major vessels colored by fluoroscopy
plt.figure(figsize=(10, 6))
sns.countplot(x='ca', hue='target', data=heart_data)
plt.title('Distribution of Number of Major Vessels Colored by Fluoroscopy')
plt.xlabel('Number of Major Vessels Colored by Fluoroscopy')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Distribution of thalassemia types
plt.figure(figsize=(10, 6))
sns.countplot(x='thal', hue='target', data=heart_data)
plt.title('Distribution of Thalassemia Types')
plt.xlabel('Thalassemia Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Calculate correlation coefficients between attributes and target variable
correlation = heart_data.corr()['target'].sort_values(ascending=False)
print("Correlation Coefficients with Target Variable (Heart Disease):")
print(correlation)

# Visualize relationships using pair plot
sns.pairplot(heart_data, hue='target', diag_kind='kde', palette='husl')
plt.show()

# Correlation matrix with target variable
plt.figure(figsize=(12, 10))
sns.heatmap(heart_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix with Target Variable (Heart Disease)')
plt.show()
```

```python
# Calculate mean values of numerical attributes by target (heart disease presence)
mean_values_by_target = heart_data.groupby('target').mean()
print("Mean Values of Numerical Attributes by Target (Heart Disease Presence):")
print(mean_values_by_target)

# Visualize distribution of age by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=heart_data)
plt.title('Distribution of Age by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('Age')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Visualize distribution of cholesterol by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=heart_data)
plt.title('Distribution of Cholesterol by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('Cholesterol')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Visualize distribution of maximum heart rate achieved by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='thalach', data=heart_data)
plt.title('Distribution of Maximum Heart Rate Achieved by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('Maximum Heart Rate Achieved')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Visualize distribution of ST depression induced by exercise relative to rest by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='oldpeak', data=heart_data)
plt.title('Distribution of ST Depression Induced by Exercise Relative to Rest by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('ST Depression')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Visualize distribution of resting blood pressure by target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='trestbps', data=heart_data)
plt.title('Distribution of Resting Blood Pressure by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('Resting Blood Pressure')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Calculate mean values of numerical attributes by target (heart disease presence)
mean_values_by_target = heart_data.groupby('target').mean()
print("Mean Values of Numerical Attributes by Target (Heart Disease Presence):")
print(mean_values_by_target)

# Calculate median values of numerical attributes by target (heart disease presence)
median_values_by_target = heart_data.groupby('target').median()
print("\nMedian Values of Numerical Attributes by Target (Heart Disease Presence):")
print(median_values_by_target)

# Calculate standard deviations of numerical attributes by target (heart disease presence)
std_values_by_target = heart_data.groupby('target').std()
print("\nStandard Deviations of Numerical Attributes by Target (Heart Disease Presence):")
print(std_values_by_target)

# Calculate correlation coefficients between attributes and target variable
correlation = heart_data.corr()['target'].sort_values(ascending=False)
print("\nCorrelation Coefficients with Target Variable (Heart Disease):")
print(correlation)
```

```python
# Count the number of individuals with and without heart disease
heart_disease_counts = heart_data['target'].value_counts()
print("\nCounts of Heart Disease Presence:")
print(heart_disease_counts)

# Calculate the percentage of individuals with heart disease
percentage_heart_disease = (heart_disease_counts[1] / len(heart_data)) * 100
print("\nPercentage of Individuals with Heart Disease: {:.2f}%".format(percentage_heart_disease))

# Calculate age distribution by heart disease presence
age_distribution_by_target = heart_data.groupby('target')['age'].describe()
print("\nAge Distribution by Heart Disease Presence:")
print(age_distribution_by_target)

# Calculate gender distribution by heart disease presence
gender_distribution_by_target = heart_data.groupby(['target', 'sex']).size().unstack(fill_value=0)
gender_distribution_by_target.columns = ['Female', 'Male']
print("\nGender Distribution by Heart Disease Presence:")
print(gender_distribution_by_target)

# Calculate chest pain type distribution by heart disease presence
chest_pain_distribution_by_target = heart_data.groupby(['target', 'cp']).size().unstack(fill_value=0)
print("\nChest Pain Type Distribution by Heart Disease Presence:")
print(chest_pain_distribution_by_target)

# Calculate correlation between selected attributes
selected_attributes = ['age', 'chol', 'thalach', 'oldpeak', 'trestbps']
correlation_selected_attributes = heart_data[selected_attributes].corr()
print("\nCorrelation between Selected Attributes:")
print(correlation_selected_attributes)
```

```python
# Calculate the mean age of individuals with and without heart disease
mean_age_by_target = heart_data.groupby('target')['age'].mean()
print("\nMean Age by Heart Disease Presence:")
print(mean_age_by_target)

# Calculate the mode of chest pain type by heart disease presence
mode_chest_pain_by_target = heart_data.groupby('target')['cp'].apply(lambda x: x.mode()[0])
print("\nMode Chest Pain Type by Heart Disease Presence:")
print(mode_chest_pain_by_target)

# Calculate the average cholesterol level of individuals with and without heart disease
average_cholesterol_by_target = heart_data.groupby('target')['chol'].mean()
print("\nAverage Cholesterol Level by Heart Disease Presence:")
print(average_cholesterol_by_target)

# Calculate the median maximum heart rate achieved by individuals with and without heart disease
median_max_heart_rate_by_target = heart_data.groupby('target')['thalach'].median()
print("\nMedian Maximum Heart Rate Achieved by Heart Disease Presence:")
print(median_max_heart_rate_by_target)

# Calculate the mean ST depression induced by exercise relative to rest by heart disease presence
mean_st_depression_by_target = heart_data.groupby('target')['oldpeak'].mean()
print("\nMean ST Depression Induced by Exercise Relative to Rest by Heart Disease Presence:")
print(mean_st_depression_by_target)

# Calculate the mode of resting blood pressure by heart disease presence
mode_resting_bp_by_target = heart_data.groupby('target')['trestbps'].apply(lambda x: x.mode()[0])
print("\nMode Resting Blood Pressure by Heart Disease Presence:")
print(mode_resting_bp_by_target)
```

```python
# Calculate the range of age for individuals with and without heart disease
age_range_by_target = heart_data.groupby('target')['age'].apply(lambda x: x.max() - x.min())
print("\nAge Range by Heart Disease Presence:")
print(age_range_by_target)

# Calculate the proportion of individuals with heart disease by gender
heart_disease_proportion_by_gender = heart_data.groupby('sex')['target'].mean()
print("\nProportion of Individuals with Heart Disease by Gender:")
print(heart_disease_proportion_by_gender)

# Calculate the distribution of thalassemia types by heart disease presence
thal_distribution_by_target = heart_data.groupby(['target', 'thal']).size().unstack(fill_value=0)
print("\nThalassemia Type Distribution by Heart Disease Presence:")
print(thal_distribution_by_target)

# Calculate the percentage of individuals with exercise-induced angina by heart disease presence
exang_percentage_by_target = heart_data.groupby('target')['exang'].mean() * 100
print("\nPercentage of Individuals with Exercise-Induced Angina by Heart Disease Presence:")
print(exang_percentage_by_target)

# Calculate the distribution of number of major vessels colored by fluoroscopy by heart disease presence
ca_distribution_by_target = heart_data.groupby(['target', 'ca']).size().unstack(fill_value=0)
print("\nNumber of Major Vessels Colored by Fluoroscopy Distribution by Heart Disease Presence:")
print(ca_distribution_by_target)
```

```python
# Calculate the frequency of different chest pain types by heart disease presence
chest_pain_frequency_by_target = heart_data.groupby('target')['cp'].value_counts(normalize=True)
print("\nFrequency of Chest Pain Types by Heart Disease Presence:")
print(chest_pain_frequency_by_target)

# Calculate the average maximum heart rate achieved by gender
average_max_heart_rate_by_gender = heart_data.groupby('sex')['thalach'].mean()
print("\nAverage Maximum Heart Rate Achieved by Gender:")
print(average_max_heart_rate_by_gender)

# Calculate the distribution of individuals with heart disease based on fasting blood sugar levels
fbs_distribution_by_target = heart_data.groupby('target')['fbs'].value_counts(normalize=True)
print("\nDistribution of Heart Disease by Fasting Blood Sugar Levels:")
print(fbs_distribution_by_target)

# Calculate the percentage of individuals with heart disease based on exercise-induced ST depression
st_depression_distribution_by_target = heart_data.groupby('target')['oldpeak'].apply(lambda x: (x > 0).mean() * 100)
print("\nPercentage of Individuals with Heart Disease by Exercise-Induced ST Depression:")
print(st_depression_distribution_by_target)

# Calculate the distribution of individuals with heart disease based on resting electrocardiographic results
restecg_distribution_by_target = heart_data.groupby('target')['restecg'].value_counts(normalize=True)
print("\nDistribution of Heart Disease by Resting Electrocardiographic Results:")
print(restecg_distribution_by_target)
```

```python
# Calculate the proportion of individuals with heart disease for each chest pain type
cp_proportion_by_target = heart_data.groupby('cp')['target'].mean()
print("\nProportion of Individuals with Heart Disease by Chest Pain Type:")
print(cp_proportion_by_target)

# Calculate the distribution of individuals with heart disease based on exercise-induced angina
exang_distribution_by_target = heart_data.groupby('target')['exang'].value_counts(normalize=True)
print("\nDistribution of Heart Disease by Exercise-Induced Angina:")
print(exang_distribution_by_target)

# Calculate the percentage of individuals with heart disease for each thalassemia type
thal_proportion_by_target = heart_data.groupby('thal')['target'].mean()
print("\nPercentage of Individuals with Heart Disease by Thalassemia Type:")
print(thal_proportion_by_target)

# Calculate the distribution of heart disease based on slope of the peak exercise ST segment
slope_distribution_by_target = heart_data.groupby('target')['slope'].value_counts(normalize=True)
print("\nDistribution of Heart Disease by Slope of the Peak Exercise ST Segment:")
print(slope_distribution_by_target)

# Calculate the proportion of individuals with heart disease based on number of major vessels colored by fluoroscopy
ca_proportion_by_target = heart_data.groupby('ca')['target'].mean()
print("\nProportion of Individuals with Heart Disease by Number of Major Vessels Colored by Fluoroscopy:")
print(ca_proportion_by_target)
```

```python
# Calculate average age by heart disease presence
average_age = heart_data.groupby('target')['age'].mean()

# Calculate proportion of individuals with heart disease by gender
gender_proportion = heart_data.groupby(['sex', 'target']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

# Calculate percentage of individuals with heart disease by chest pain type
cp_percentage = heart_data.groupby('cp')['target'].mean() * 100

# Calculate percentage of individuals with heart disease by fasting blood sugar levels
fbs_percentage = heart_data.groupby('fbs')['target'].mean() * 100

# Calculate percentage of individuals with heart disease by thalassemia type
thal_percentage = heart_data.groupby('thal')['target'].mean() * 100

# Print the results
print("Average Age by Heart Disease Presence:")
print(average_age)

print("\nProportion of Individuals with Heart Disease by Gender:")
print(gender_proportion)

print("\nPercentage of Individuals with Heart Disease by Chest Pain Type:")
print(cp_percentage)

print("\nPercentage of Individuals with Heart Disease by Fasting Blood Sugar Levels:")
print(fbs_percentage)

print("\nPercentage of Individuals with Heart Disease by Thalassemia Type:")
print(thal_percentage)
```

```python
import matplotlib.pyplot as plt

# Plot pie chart for gender distribution of individuals with heart disease
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
labels = ['Female', 'Male']
colors = ['#ff9999','#66b3ff']
plt.pie(gender_proportion.loc[:, 1], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution of Individuals with Heart Disease')

# Plot pie chart for distribution of heart disease by chest pain type
plt.subplot(1, 2, 2)
labels = ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(cp_percentage, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Heart Disease by Chest Pain Type')

plt.tight_layout()
plt.show()
```

```python
# Plot line graph for percentage of individuals with heart disease by chest pain type
plt.figure(figsize=(10, 5))
plt.plot(cp_percentage.index, cp_percentage.values, marker='o', color='b', linestyle='-')
plt.title('Percentage of Individuals with Heart Disease by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Percentage with Heart Disease (%)')
plt.xticks(cp_percentage.index, ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'])
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Path to the text file
text_file_path = '/content/Dataset Details.txt'

# Read the text file into a list of lines
with open(text_file_path, 'r') as file:
    lines = file.readlines()

# Parse the lines to create a list of dictionaries
data = []
for line in lines:
    fields = line.strip().split(',')  # Assuming comma-separated fields
    if len(fields) >= 5:  # Skip lines with insufficient fields
        record = {
            'Dataset Name': fields[0],
            'Description': fields[1],
            'Format': fields[2],
            'Size': fields[3],
            'Link': fields[4]
        }
        data.append(record)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data)

# Path to save the CSV file
csv_file_path = 'dataset_details.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)

print("CSV file successfully created!")
```
